```python
# We are basically working with text data 

# df.str.lower(), df.str.upper(),s.str.len()

# https://pandas.pydata.org/pandas-docs/stable/user_guide/text.html
```

```python
pub = data['Place of Publication']
london = pub.str.contains('London')
oxford = pub.str.contains('Oxford')
```

```python
data['Place of Publication'] = np.where(london, 'London',
                                      np.where(oxford, 'Oxford',
                                               pub.str.replace('-', ' ')))
```

```python
data['Place of Publication'].unique().tolist()
```

```python
# COLUMN 2
data['Date of Publication'] = data['Date of Publication'].str.extract(r'^(\d{4})', expand=False)

'''
1879 [1878]
^ start here
\d start of the digit
{4} repat the above rule four times
'''
```
