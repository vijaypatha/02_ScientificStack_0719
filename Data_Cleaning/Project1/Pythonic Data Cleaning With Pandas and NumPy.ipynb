```python
# Date of publication multiple years etc. Lets clean up date of publication 

'''
The \d represents any digit, and {4} repeats this rule four times. 
The ^ character matches the start of a string, and the parentheses denote a capturing group, 
which signals to Pandas that we want to extract that part of the regex.
(We want ^ to avoid cases where [ starts off the string.)
'''

extr = df['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
```

```python
df['Date of Publication'] = pd.to_numeric(extr)
df.dtypes
```

```python
pub = df['Place of Publication']
london = pub.str.contains('London')
london[:10]
```

```python
oxford = pub.str.contains('Oxford')
oxford[0:10]
```

```python
'''
Intuitively, np.where is like asking "tell me where in this array, entries satisfy a given condition".

np.where(condition1, x1, 
        np.where(condition2, x2, 
            np.where(condition3, x3, ...)))

'''
```

```python
df['Place of Publication'] = np.where(london, 'London',
                                      np.where(oxford, 'Oxford',
                                               pub.str.replace('-', ' ')))
df['Place of Publication'].head(15)
```
